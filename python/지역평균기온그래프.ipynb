```python
fp = open('/content/drive/MyDrive/지역평균기온 (1).txt')
data = fp.readlines()
fp.close()
data
```

```python
dt = {} #빈 딕셔너리 생성

for line in data:
  line = line.replace("\n", "")
  #print(line)
  items = line.split(',')
  #print(items)
  dt[items[0]] = int(items[1])

print(type(dt))
dt
```

```python
x = list(dt.keys())
y = list(dt.values())

print(x)
print(y)
```

```python
plt.plot(x, y, 'go-', color='pink')

plt.title('지역평균기온')
plt.xlabel('도시')
plt.ylabel('온도')

for i in range(0, len(x)):
  plt.text(i, y[i], y[i], ha = 'center')


plt.show()
```
