##**데이터 시각화하기**

###① 객실 등급(Pclass) 속성과 생존율 관계 분석 및 시각화하기
- 나의 예측 : 좋은 객실에 있는 사람들이 잘 살았을 것 같다. 객식등급, 생존률 상관관계 성립.

####- 타이태닉 데이터의 객실 등급별 다른 속성들의 평균값 비교

```python
#데이터 속성의 그룹별 통계를 구할 수 있는 groupby() 함수와 값
data.groupby('Pclass').mean()
```

####- 타이태닉 데이터의 객실 등급별 생존율 평균 시각화

```python
# plot() 함수와 kin='bar' 옵션, ['속성 이름']을 이용하여 막대그래프로 시각화 (객식등급, 생존률 상관관계 성립)
data.groupby('Pclass').mean() ['Survived'].plot(kind = 'bar')
```

```python
data.groupby('Age').mean() ['Survived'].plot(kind = 'bar')
```

```python
#원하는 데이터를 뽑아서 데이터프레임에 필드 추가 시키기
data['Age_range'] = pd.cut(data['Age'], bins = [0, 7, 15, 35, 60, 100],
                           include_lowest=True,
                           labels=['baby', 'teen', 'young', 'adult', 'old'])

data
```

```python
data.groupby('Age_range').mean() ['Survived'].plot(kind = 'bar')
```

```python
data.groupby('Age_range').count().plot(kind = 'bar')
```

```python
subset = data[['Age', 'Survived']]
subset
```
